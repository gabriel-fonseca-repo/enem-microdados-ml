# Gradiente

No cálculo vetorial, o gradiente de uma função diferenciável escalar $f$ de várias variáveis ​​é o campo vetorial $\nabla f$ cujo valor em um ponto $p$ é o vetor cujos componentes são as derivadas parciais de $f$ em $p$.

Ou seja, para $f : R^n \rightarrow R$, seu gradiente $∇ f : R^n \rightarrow R^n$ é definido no ponto $p = ( x_1, \ldots, x_n)$ no espaço n-dimensional como o vetor:

$\nabla f(p)= \begin{bmatrix} \frac{\partial f}{\partial x_1}(p) \\ \vdots \\ \frac{\partial f}{\partial x_{n}}(p) \end{bmatrix}$

## Exemplo 1

```python
from mpl_toolkits import mplot3d
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt

from IPython.display import clear_output
import ipywidgets as widgets
import numpy as np

def f(x, y):

    return (x ** 2 + y ** 2)

x = np.linspace(-1, 1, 100)

y = np.linspace(-1, 1, 100)

X, Y = np.meshgrid(x, y)

Z = f(X, Y)

fig = plt.figure(figsize=(8,6))

ax = fig.add_subplot(projection='3d')
ax.view_init(elev=40., azim=45)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
              cmap='viridis', edgecolor='none')
ax.set_xlabel('x')
ax.set_ylabel('y')

plt.show()
```

```python
# gradiente
def gx(x, y):
    return 2 * x

def gy(x, y):
    return 2 * y

x = np.linspace(-1, 1, 10)
y = np.linspace(-1, 1, 10)

X1, Y1 = np.meshgrid(x, y)
Gx = gx(X1, Y1)
Gy = gy(X1, Y1)
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig,ax=plt.subplots(1,1,figsize=(8,6))
cp = ax.contour(X, Y, Z,cmap='viridis',levels=5)
fig.colorbar(cp)
ax.clabel(cp,cp.levels,fmt='%.2f')

ax.quiver(X1,Y1,Gx,Gy)

ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```

# Descida em Gradiente

Começa-se com um palpite x0 para um mínimo local de F, e considera-se a sequência x0 ,x1 ,x2 ,... tal que

$x_{n + 1} = x_n - \gamma_n \nabla F(x_n),\quad n \ge 0.$

Temos uma sequência monotônica

$F(x_0) \le F (x_1) \le F(x_2) \le \ldots,$

Assim, esperamos que a sequência (x_n) convirja para o mínimo local desejado.

```python
from mpl_toolkits import mplot3d
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt

from IPython.display import clear_output
import ipywidgets as widgets
import numpy as np

def f(x, y):
    return np.sin(x) ** 2 + np.cos(5 + x * y) + 2 * np.cos(x)
x = np.linspace(0, 5, 100)
y = np.linspace(0, 5, 100)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure(figsize=(8,6))

ax = fig.add_subplot(projection='3d')
ax.view_init(elev=40., azim=45)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
              cmap='viridis', edgecolor='none')
ax.set_xlabel('x')
ax.set_ylabel('y')

plt.show()
```

```python
def gx(x, y):
    return -2.0*np.sin(x) + np.sin(2*x) - y*np.sin(5 + x*y)

def gy(x, y):
    return -x*np.sin(5.0 + x*y)

p = np.array([2,4])
gamma = 0.1

tr =[]
for i in range(1000):
  fg = np.array([gx(p[0],p[1]),gy(p[0],p[1])])
  p = p - gamma*fg
  # print(p)
  tr.append(p)

tr = np.array(tr)

print(tr.shape)
```

```python
print(tr)
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig,ax=plt.subplots(1,1,figsize=(12,10))
cp = ax.contour(X, Y, Z,cmap='viridis',levels=20)
fig.colorbar(cp)

ax.plot(tr.T[0],tr.T[1],color='red',linewidth=2.0)

ax.set_xlabel('intercept')
ax.set_ylabel('theta')
plt.show()
```
